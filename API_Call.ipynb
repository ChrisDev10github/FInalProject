```python
import requests
import os
import json
import pandas as pd
```

```python
url = "https://data.sba.gov/api/3/action/datastore_search?resource_id=dc66702f-7eff-4ec0-b998-550ac063ab78&offset="
yearID = ['dc66702f-7eff-4ec0-b998-550ac063ab78','84741c22-bdee-4a0e-b5aa-6a982a85f10b','57cf983a-37e6-495f-80d3-871bef81ad60','7663d46d-09dd-4e54-8542-bbae5f6d3386','553a423b-63ba-4810-9ee3-0ba9f589a6b3','1d3ce89c-0518-4b24-8173-79db32d70de3','41625060-a76b-4608-ad63-d969f0a4bc6f','9b355f56-aa00-462d-9e46-c2f2101d4900']
loans = []
for i in range(0,len(yearID)):
    x=True
    num=0
    while x:
        tempURL = f"https://data.sba.gov/api/3/action/datastore_search?resource_id={yearID[i]}&offset={str(num)}"
        response = requests.get(tempURL)

        j = response.json()
        j = j['result']
        data = j['records']

        if len(data) < 100:
            x = False

        num = num + 100
        loans.append(data)
    
print(len(loans))
```

```python
flatten_list = []
for s in loans:
    for i in s:
        flatten_list.append(i)
len(flatten_list)
flatten_list[3000]
```

```python
df = pd.DataFrame.from_records(flatten_list)
```
